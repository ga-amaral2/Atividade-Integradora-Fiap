# Relatório Operacional de Pré-Decolagem (Atividade Integradora)
Um programa para interpretar dados de telemetria em uma planilha e decidir se a nave está pronta para decolar ou se a decolagem deve ser abortada.

# 1. Definir as "Faixas Seguras" (Parâmetros de Voo)

````python
MIN_TEMP_INTERNA = 18
MAX_TEMP_INTERNA = 27

MIN_TEMP_EXTERNA = -20
MAX_TEMP_EXTERNA = 40

OK_INTEGRIDADE_ESTRUTURAL = 1
FALHA_INTEGRIDADE_ESTRUTURAL = 0

MIN_NIVEL_ENERGIA = 80

MIN_PRESSAO_TANQUES = 50
MAX_PRESSAO_TANQUES = 120

OK_MODULOS_CRITICOS = 1
FALHA_MODULOS_CRITICOS = 0
````

# 2. Simular a Leitura dos Dados (Telemetria)

##a) Leitura de um conjunto de dados inseridos manualmente

````python
import sys

dados = 0

while True:
    try:
        match dados:
          case 0:
            temperatura_interna = float(input(f"Digite a temperatura interna em Celsius (>= {MIN_TEMP_INTERNA} e <= {MAX_TEMP_INTERNA}): " ))
          case 1:
            temperatura_externa = float(input(f"\nDigite a temperatura externa em Celsius (>= {MIN_TEMP_EXTERNA} e <= {MAX_TEMP_EXTERNA}): " ))
          case 2:
            integridade_estrutural = int(input(f"\nIntegridade estrutural ({OK_INTEGRIDADE_ESTRUTURAL} = OK e {FALHA_INTEGRIDADE_ESTRUTURAL} = Falha): " ))
            if (integridade_estrutural != OK_INTEGRIDADE_ESTRUTURAL) and (integridade_estrutural != FALHA_INTEGRIDADE_ESTRUTURAL):
              print(f"\nErro! Por favor, digitar apenas {OK_INTEGRIDADE_ESTRUTURAL} ou {FALHA_INTEGRIDADE_ESTRUTURAL}!")
              raise RuntimeError()
          case 3:
            nivel_energia = input(f"\nNível de energia (>= {MIN_NIVEL_ENERGIA}%): " )
            if ("%" in nivel_energia):
              nivel_energia = nivel_energia.replace("%", "")
            nivel_energia = float(nivel_energia)
            if nivel_energia<0 or nivel_energia>100:
              print("\nErro! Por favor, digitar apenas números entre 0 ou 100!")
              raise RuntimeError()
          case 4:
            pressao_tanques = float(input(f"\nPressão dos tanques em bar (>={MIN_PRESSAO_TANQUES} e <={MAX_PRESSAO_TANQUES}): " ))
          case 5:
            modulos_criticos = int(input(f"\nStatus dos módulos críticos ({OK_MODULOS_CRITICOS} = OK e {FALHA_MODULOS_CRITICOS} = Falha): " ))
            if (modulos_criticos != OK_MODULOS_CRITICOS) and (modulos_criticos != FALHA_MODULOS_CRITICOS):
              print(f"\nErro! Por favor, digitar apenas {OK_MODULOS_CRITICOS} ou {FALHA_MODULOS_CRITICOS}!")
              raise RuntimeError()
          case _:
            print("\nDados recebidos com sucesso!")
            break

    except ValueError:
        print("\nErro! Por favor, digitar apenas números, sem letras ou caracteres especiais!")
        print("Vamos tentar de novo...\n")
    except RuntimeError:
        print("Vamos tentar de novo...\n")
    else:
        dados += 1
````
